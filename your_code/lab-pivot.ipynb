#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1
# your answer here
employee.groupby(['Department']).agg({'Salary':'mean'})
```

```python
employee.pivot_table(index=['Department'], values = ['Salary'], aggfunc = ['mean'])
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# Method 2
# your answer here
employee.groupby(['Department', 'Title']).agg({'Salary': 'mean'})
```

```python
# your answer here
employee.pivot_table(index = ['Department', 'Title'], values = ['Salary'], aggfunc = ['mean'])
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.

```python
# your answer here
employee.pivot_table(index = ['Title'], values = ['Salary', 'Years'], aggfunc = {'Salary':'mean', 'Years':'sum'})
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# your answer here
    
employee.pivot_table(index = ['Department', 'Title'], values = ['Salary', 'Name'], aggfunc = {'Salary':'mean', 'Name':'count'})
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
STD_ = employee.pivot_table(index = ['Department', 'Title'], values = ['Salary'], aggfunc = ['mean', min, max, 'std'])

'''We get NaN values for those rows where there are only 1 employee, so we can't calculate STD. 
For HR Analyst and IT Analyst we found a small deviation, and for Sales Associate both employees earn the same amount.'''
STD_
```
